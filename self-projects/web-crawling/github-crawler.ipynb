### Simple Web Crawler – Page Title Extraction

The crawler starts from a given **seed URL**, visits web pages one by one, extracts the **title of each page**, and follows the links found on the page.  
The process continues until a fixed number of pages are crawled.

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
```

```python
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Educational crawler for IR course)"
}
```

```python
START_URL = "https://github.com/manishtiwari-45"

MAX_PAGES = 50
OUTPUT_FILE = "github_titles.txt"
```

```python
frontier = [START_URL]   
visited = set()          
```

```python
output = open(OUTPUT_FILE, "w", encoding="utf-8")
print("Output file opened:", OUTPUT_FILE)
```

```python
while frontier and len(visited) < MAX_PAGES:
    url = frontier.pop(0)

    if url in visited:
        continue

    print(f"\nCrawling ({len(visited)+1}/{MAX_PAGES}):", url)
    visited.add(url)

    try:
        response = requests.get(url, headers=HEADERS)
    except:
        print("  Failed to fetch page")
        continue


    soup = BeautifulSoup(response.text, "html.parser")

    if soup.title:
        title = soup.title.get_text().strip()
    else:
        title = "No Title"

    print("  Title:", title)

    output.write(url + "\n")
    output.write(title + "\n\n")


    links = soup.find_all("a")

    for link in links:
        href = link.get("href")
        if not href:
            continue

        absolute_url = urljoin(url, href)

        if (
            absolute_url not in visited
            and absolute_url not in frontier
        ):
            frontier.append(absolute_url)

    print("  Frontier size:", len(frontier))


output.close()
print("\nDone!")
print("Total pages crawled:", len(visited))
print("Titles saved in:", OUTPUT_FILE)
```
