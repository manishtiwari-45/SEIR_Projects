```python
import os

start_path = os.path.expanduser("~")
```

```python
all_files = []

for root, dirs, files in os.walk(start_path):
    for file in files:
        file_path = os.path.join(root, file)
        all_files.append(file_path)
```

```python
print("Total files found:", len(all_files))
```

```python
for file in all_files[:20]:
    print(file)
```
